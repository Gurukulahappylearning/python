# flowchart of while loop

```python
# print hai for 5 times
print("Hai")
print("Hai")
print("Hai")
print("Hai")
print("Hai")
```

```python
# using while
i=1
while(i<=5):
    print("Hai")
    i=i+1
```

```python
# example to print numbers without using loop
print("number 1")
print("number 2")
print("number 3")
print("number 4")
print("number 5")
print("number 6")
print("number 7")
print("number 8")
print("number 9")
print("number 10")
```

```python
# example printing numbers using loop
i=1
while(i<=10):
    print("Number",i)
    i=i+1
print("terminated from while loop")
    
```

```python
# even numbers
i=2
while(i<=20):
    print(i)
    i=i+2
print("program terminated")
```

```python
i=20
while(i>=2):
    print(i)
    i=i-2
print("program terminated")
```

```python
# infinite loop
i=1
while(i<=10):
    print("Number",i)

print("terminated from while loop")
```

```python
# using true keyword
i=1
while True:
    print("Number ",i)
```
